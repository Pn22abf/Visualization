```python
#Pandas :we use pandas to read csv file
# Numpy :It is widely used in scientific computing, data analysis, and machine learning applications.
#LabelEncoder :  is used to convert categorical or text data into numerical data. 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

```python
# We can use funtion to show some specific columns from the dataset

def show_specific_columns(csv_path, columns):
    # Load CSV data into a pandas dataframe
    df = pd.read_csv('USA_cars_datasets.csv')
    
    # Select only the specified columns
    df = df[columns]
    
    return df
```

```python
csv_path = 'USA_cars_datasets.csv'
columns = ['price', 'model', 'state']
unique_df = show_specific_columns(csv_path, columns)
print(unique_df)
```

```python
import csv

def get_column(file_path, column_name):
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        col_value = [row[column_name] for row in csv_reader]
    return col_value
```

```python
col_val = get_column("USA_cars_datasets.csv", "brand")
print(col_val)
```

```python
#A histogram is a graphical representation of the distribution of a dataset
#.title() is a method provided by many Python plotting libraries, including Matplotlib and Seaborn, 
#that is used to set the title of a plo
plt.hist(df['price'], color = 'black', edgecolor = 'blue',bins=10)


plt.title(' Types of Car Models')
plt.xlabel('Car Models')
plt.ylabel('Number of cars')
```
